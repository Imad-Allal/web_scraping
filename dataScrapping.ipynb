```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://fr.wikipedia.org/wiki/Alger')
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
paragraphs = soup.find_all('span', attrs={'class':'mw-headline'})
```

```python
for paragraph in paragraphs:
  print(paragraph.text)
```

```python
paragraphs = soup.find_all('table', attrs={'class':'infobox_v2 noarchive'})
print(paragraphs)
```

```python
for p in paragraphs:
  if p == '<th scope="row" style="vertical-align: bottom"><a href="/wiki/Densit%C3%A9_de_population" title="Densité de population">Densité</a></th>'
  print(p)
```
